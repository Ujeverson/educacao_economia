```python
import pandas as pd
```

```python
arquivo ='muni_exp.csv'
#http://comexstat.mdic.gov.br/pt/municipio
```

```python
df = pd.read_csv(arquivo, sep=';')
df
```

```python
# Remover a string ' - GO' da coluna 'Município'
df['Município'] = df['Município'].str.replace(' - GO', '').fillna(df['Município'])
df = df.drop(['UF do Município', 'Codigo SH4'], axis=1)
df
```

```python
# Ordenar o DataFrame pelo ano
df = df.sort_values(by='Ano')
df = df.reset_index(drop=True)
df
```

```python
df.columns = ['ANO','NMU','MCEXT','CTRY','US$']
df.head(15)
```

```python
# Agrupar os valores por 'ANO' e 'NMU', juntando os valores da coluna 'MCEXT' e 'CTRY', e somando os valores da coluna 'US$'
df1 = df.groupby(['ANO', 'NMU'], as_index=False).agg({'MCEXT': ' '.join, 'CTRY': ' '.join, 'US$': 'sum'})
df1.head(15)
```

```python
# Trocar a ordem das colunas 0 e 1
colunas = list(df1.columns)
colunas[0], colunas[1] = colunas[1], colunas[0]
df1 = df1[colunas]
df1
```

```python
df1.to_csv('export.csv', index=False)
```
