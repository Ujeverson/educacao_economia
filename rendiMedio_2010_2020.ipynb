```python
import pandas as pd
```

```python
arquivo = "rendimento_salario.csv"
```

```python
df = pd.read_csv(arquivo, sep=";")

df = df.replace({' -':'0', '-':'0', '   ':'0', '    ':'0',' -   ':'0','.':'',',':'.'}).fillna(0)
df
```

```python
dic = {
    'Localidade':'NMU',
    'Ano': 'ANO',
    'Variável':'VAR',
    'Rendimento Médio (R$)': 'RM',
    'Rendimento Médio - Extrativa Mineral (R$)': 'REM',
    'Rendimento Médio - Indústria de Transformação (R$)': 'RIT',
    'Rendimento Médio - Serviços Industriais de Utilidade Pública (R$)': 'RSIUP',
    'Rendimento Médio - Construção Civil (R$)': 'RCC',
    'Rendimento Médio - Comércio (R$)': 'RCO',
    'Rendimento Médio - Serviços (R$)': 'RSE',
    'Rendimento Médio - Administração Pública (R$)': 'RAP',
    'Rendimento Médio - Agropecuária, Extração Vegetal, Caça e Pesca (R$)': 'RAECP'
}

df = df.rename(columns=dic)
df
```

```python
# Definir os tipos de dados específicos para cada coluna
df['NMU'] = df['NMU'].astype(str)
df['ANO'] = df['ANO'].astype(str)
df['REM'] = df['REM'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)
df['RIT'] = df['RIT'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)
df['RSIUP'] = df['RSIUP'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)
df['RCC'] = df['RCC'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)
df['RCO'] = df['RCO'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)
df['RSE'] = df['RSE'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)
df['RAP'] = df['RAP'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)
df['RAECP'] = df['RAECP'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)
df
```

```python
# salva o dataframe modificado
df.to_csv("trabalho.csv", index=False)
```
